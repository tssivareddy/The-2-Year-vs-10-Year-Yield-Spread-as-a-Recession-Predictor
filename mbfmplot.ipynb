```python
import pandas as pd
from google.colab import files
import io
import plotly.graph_objects as go
```

```python
def do(bond):
  uploaded = files.upload()
  file_name = next(iter(uploaded.keys()))
  xls_2 = pd.read_csv(io.BytesIO(uploaded[file_name]))
  uploaded = files.upload()
  file_name = next(iter(uploaded.keys()))
  xls_10 = pd.read_csv(io.BytesIO(uploaded[file_name]))

  df_2y = xls_2.copy()
  df_2y['Date'] = pd.to_datetime(df_2y['Date'])
  df_2y.set_index('Date', inplace=True)

  df_10y = xls_10.copy()
  df_10y['Date'] = pd.to_datetime(df_10y['Date'])
  df_10y.set_index('Date', inplace=True)

  df = pd.merge(df_2y, df_10y, left_index=True, right_index=True, how='inner')

  df['Yield_Spread'] = df[' Close_y'] - df[' Close_x']

  fig = go.Figure()

  fig.add_trace(go.Scatter(
      x=df.index,
      y=df['Yield_Spread'],
      mode='lines',
      name='10Y - 2Y Yield Spread',
      line=dict(color='blue')
  ))

  fig.add_trace(go.Scatter(
      x=[df.index.min(), df.index.max()],
      y=[0, 0],
      mode='lines',
      name='Inversion Threshold',
      line=dict(color='red', dash='dash')
  ))

  fig.update_layout(
      title=bond+"2-Year vs. 10-Year Yield Spread",
      xaxis_title="Date",
      yaxis_title="Yield Spread (10Y - 2Y)",
      template="plotly_white",
      legend=dict(x=0.02, y=0.98),
  )

  fig.show()




```
